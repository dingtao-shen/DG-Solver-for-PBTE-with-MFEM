```python
import numpy as np
import quadpy
```

```python
for degree in range(1, 15):
    scheme = quadpy.t3.get_good_scheme(degree)
    points = scheme.points
    weights = scheme.weights
    data = np.column_stack((points.T, weights.T))

    np.savetxt('quad3D_'+str(degree)+'.txt', 
            data, 
            fmt='%.12f',
            delimiter=' ', 
            header='',
            comments='')
```
